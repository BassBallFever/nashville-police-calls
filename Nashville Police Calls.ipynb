# Read data
Data was downloaded from data.nahsville.gov

```python
calls = pd.read_csv('Metro_Nashville_Police_Department_Calls_for_Service.csv')
```

# EDA
Start with checking the number of rows and columns and look at the first 5 rows, then check the types of each column.

```python
print(calls.shape)
calls.head()
```

```python
calls.dtypes
```

## Find first and last call
The Call Received column contains date and time information, but is an object. Find the first and last based on the string values contained in Call Received.

```python
firstcalls = calls[calls['Call Received'].str.contains('01/01/2013 12:00')]
firstcalls
```

```python
lastcalls = calls[calls['Call Received'].str.contains('11/10/2020 11:59')]
lastcalls
```

## Inspect null values

```python
calls.isnull().sum()/calls.shape[0]*100
```

## Drop columns with all null values

```python
calls = calls.drop(['Shift','Disposition Description'], axis=1)
```

## Inspect calls that have null values for Unit Dispatched

```python
noUnit = calls[calls['Unit Dispatched'].isnull()]
noUnit.head()
```

```python
print(noUnit.shape)
noUnit.isnull().sum()
```

```python
noUnit['Disposition Code'].value_counts()
```

## Results of Disposition Code
The Disposition Code shows that almost 85% of the calls with no Unit Dispatched were disregarded, report completed, or assisted citizen. Most of the nulls can be attributed to no units being dispatched for those calls.

## Inspect calls with location data available
Check the calls that have location data availabe in order to determine if we can use that data.

```python
locationCalls = calls[calls['Latitude'].notnull()]
locationCalls.head()
```

```python
locationCalls.shape
```

```python
location2013 = locationCalls[locationCalls['Call Received'].str.contains('2013')]
location2020 = locationCalls[locationCalls['Call Received'].str.contains('2020')]
```

```python
print('Location in 2013: {0}'.format(location2013.shape[0]))
print('Location in 2020: {0}'.format(location2020.shape[0]))
```
